# C2.7 VISUALIZING GENDER DISTRIBUTION FOR EACH COMPANY

```python
gender_count = main_df.groupby(['Company', 'Gender']).size().unstack()

# Create a pie chart for each company
fig, axes = plt.subplots(nrows=1, ncols=len(gender_count), figsize=(10, 5))

for i, (company, data) in enumerate(gender_count.iterrows()):
    ax = axes[i]
    ax.pie(data, labels=data.index, autopct='%1.1f%%', startangle=90)
    ax.set_title(company)

plt.suptitle('Gender Distribution by Company')
plt.tight_layout()
plt.show()
```

# C2.8 VISUALIZING PERCENTAGE OF USERS IN EACH CITY

```python
grouped = main_df.groupby(['City', 'Company']).size().unstack()
grouped_percentage = grouped.div(grouped.sum()) * 100
colors = ['pink', 'yellow']
plt.figure(figsize=(16, 9))
bars = grouped_percentage.plot(kind='bar', stacked=False, width=1, color=colors)


for bar in bars.containers:
    for rect in bar:
        height = rect.get_height()
        percentage = f'{height:.1f}%'
        plt.text(rect.get_x() + rect.get_width() / 2, height, percentage, ha='center', va='bottom', fontsize=7)

plt.xlabel('City', fontsize=7)
plt.ylabel('Percentage of Users', fontsize=10)
plt.title('Percentage of Users by City for Different Companies', fontsize=12)
plt.legend(title='Company')
plt.xticks(rotation=90, fontsize=6)
plt.yticks(fontsize=8)
plt.tight_layout()
plt.show()
```

# 2.9 CALCULATING AND VISUALIZING THE YEARLY PROFIT FOR EACH COMPANY

```python
main_df['Profit'] = main_df['Price Charged'] - main_df['Cost of Trip']
main_df['Year'] = pd.to_datetime(main_df['Date of Travel']).dt.year

yearly_profit = main_df.groupby(['Year', 'Company'])['Profit'].sum().unstack()


colors = ['pink', 'yellow']

# Create a bar chart for yearly profit analysis
yearly_profit.plot(kind='bar', color=colors)

plt.xlabel('Year')
plt.ylabel('Profit (USD)')
plt.title('Yearly Profit Analysis by Cab Company')
plt.show()

print(yearly_profit)
```

 As expected, yellow cab makes more profit yearly

# 2.10 CALCULATING AND VISUALIZING THE MONTHLY PROFIT FOR EACH COMPANY

```python
main_df['Profit'] = main_df['Price Charged'] - main_df['Cost of Trip']

# Extract the year and month from the 'Date of Travel' column
main_df['Year'] = pd.to_datetime(main_df['Date of Travel']).dt.year
main_df['Month'] = pd.to_datetime(main_df['Date of Travel']).dt.month
monthly_profit = main_df.groupby(['Year', 'Month', 'Company'])['Profit'].sum().unstack()

colors = ['pink', 'yellow']
# Create a line chart for monthly profit analysis
monthly_profit.plot(kind='line', marker='o', color=colors)

# Add labels and title
plt.xlabel('Month')
plt.ylabel('Profit (USD)')
plt.title('Monthly Profit Analysis by Cab Company')
plt.show()
```

BOTH COMPANIES HAVE LOW PROFIT DURING THE SPRING AND LOWER DURING THE SUMMER AND VERY HIGH PROFIT DURING WINTER

# # 2.11 CALCULATING AND VISUALIZING THE AVERAGE INCOME FOR EACH USERS

```python
# Group the data by Company and calculate the average income
average_income_by_company = main_df.groupby('Company')['Income (USD/Month)'].mean()

# Prepare data for the box plot
data = [main_df[main_df['Company'] == company]['Income (USD/Month)'] for company in average_income_by_company.index]
colors = {'Pink Cab': 'pink', 'Yellow Cab': 'yellow'}


plt.figure(figsize=(10, 6))
bp = plt.boxplot(data, labels=average_income_by_company.index, patch_artist=True, notch=True)

for box, company in zip(bp['boxes'], average_income_by_company.index):
    box.set(facecolor=colors[company])

plt.xlabel('Company')
plt.ylabel('Average Income (USD/Month)')
plt.title('Average Income Range by Company')
plt.show()

print(average_income_by_company)
```

# C3.1 HYPOTHESIS 1 - IS THERE ANY DIFFERNCE IN PROFIT IN RESPECT TO SEASON (WINTER, SPRING SUMMER & FALL)

H0 : There is no or barely any difference in profit in both cab companies with respect to seasons. 

H1 : There is a marginal difference in profit in both cab companies with respect to seasons. 

```python
#Winter: January, February, December
#Summer: June, July, August
#Fall: September, October, November
main_df['Date of Travel'] = pd.to_datetime(main_df['Date of Travel'])
main_df['Month'] = main_df['Date of Travel'].dt.month

# Assign seasons based on the month
main_df['Season'] = pd.cut(main_df['Month'], bins=[1, 3, 6, 9, 12], labels=['Winter', 'Spring', 'Summer', 'Fall'])
main_df['Profit'] = main_df['Price Charged'] - main_df['Cost of Trip']
profit_by_company_season = main_df.groupby(['Company', 'Season'])['Profit'].sum().reset_index()
companies = profit_by_company_season['Company'].unique()
colors = {'Pink Cab': 'pink', 'Yellow Cab': 'yellow'}

# Create a bar chart for each company
for company in companies:
    company_data = profit_by_company_season[profit_by_company_season['Company'] == company]
    seasons = company_data['Season']
    profits = company_data['Profit']
    plt.figure(figsize=(10, 6))
    plt.bar(seasons, profits, color=colors[company])
    plt.xlabel('Season')
    plt.ylabel('Profit')
    plt.title(f'Profit for {company} in Different Seasons')
    plt.show()
seasonal_profit_by_cab = main_df.groupby(['Season', 'Company'])['Profit'].sum()

print(seasonal_profit_by_cab)
```
